## Contains Duplicates

```python
def containsDuplicate(self, nums):
    """
    Time: O(n) | Space: O(n)
    """
    collection = set()
    for num in nums:
        if num in collection: return True
        collection.add(num)
    return False
```

```python
def containsDuplicate(self, nums):
    """
    Time: O(nlog(n)) | Space: O(1)
    """
    nums.sort()
    for i in range(len(nums) - 1):
        if nums[i] == nums[i + 1]:
            return True
    return False
```

## Valid Anagram

```python
def isAnagram(self, s, t):
    """
    Time: O(n^2) | Space: O(n)
    """
    word = list(t)
    for ch in s:
        if ch not in word:
            return False
        word.remove(ch)
    return False if word else True
```

```python
def isAnagram(self, s, t):
    """
    Time: O(nlog(n)) | Space: O(1)
    """
    return sorted(s) == sorted(t)
```

## Two Sum

```python
def twoSum(self, nums, target):
    """
    Time: O(n^2) | Space: O(1)
    """
    for i in range(len(nums) - 1):
        for j in range(i + 1, len(nums)):
            if nums[i] + nums[j] == target:
                return [i, j]
```

```python
def twoSum(self, nums, target):
    """
    Time: O(n) | Space: O(n)
    """     
    hashmap = {}
    for i, firstNum in enumerate(nums):
        secondNum = target - firstNum
        if secondNum in hashmap:
            return [i, hashmap[secondNum]]
        else:
            hashmap[firstNum] = i
```
