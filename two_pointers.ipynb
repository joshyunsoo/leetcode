## Valid Palindrome

```python
def isPalindrome(self, s):
    """
    Time: O(n) | Space: O(n)
    Strategy: Check if s == s[::-1]
    """
    newStr = ""
    for i in range(len(s)):
        if s[i].isalnum():
            newStr += s[i].lower()
    return newStr == newStr[::-1]   
```

```python
def isPalindrome(self, s):
    """
    Time: O(n) | Space: O(1)  
    Strategy: Two Pointers
    """    
    left, right = 0, len(s) - 1
    
    while left < right:
        if s[left].isalnum():
            if s[right].isalnum():
                if s[left].lower() != s[right].lower(): 
                    return False
                left, right = left + 1, right - 1
            else:
                right -= 1
        else:
            left += 1
    return True
```
